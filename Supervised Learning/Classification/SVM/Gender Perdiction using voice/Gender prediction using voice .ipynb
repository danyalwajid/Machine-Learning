```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('voice.csv')
df.head()
```

```python
df.corr()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
print(f"Total Number of labels {df.shape[0]}")
print(f"Number of males : {df[df.label == 'male'].shape[0]}")
print(f"Number of females : {df[df.label == 'female'].shape[0]}")
```

```python
X = df.iloc[:,:-1]
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
y = df.iloc[:,-1]
gendar_encoder = LabelEncoder()
y = gendar_encoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test = train_test_split(X, y , test_size= 0.2, random_state= 1)
```

```python
from sklearn.svm import SVC
from sklearn import metrics 

svc = SVC()
svc.fit(X_train , y_train)
y_predict = svc.predict(X_test)
print("Accuracy Score")
print(metrics.accuracy_score(y_test, y_predict))
```

```python
svc = SVC(kernel = "linear")
svc.fit(X_train, y_train)
y_predict = svc.predict(X_test)
print("Accuracy Score")
print(metrics.accuracy_score(y_test, y_predict))
```

```python
svc = SVC(kernel = "rbf")
svc.fit(X_train, y_train)
y_predict = svc.predict(X_test)
print("Accuracy Score")
print(metrics.accuracy_score(y_test,y_predict))
```

```python
svc = SVC(kernel = "poly")
svc.fit(X_train, y_train)
y_predict = svc.predict(X_test)
print("Accuracy Score")
print(metrics.accuracy_score(y_test, y_predict))
```

```python
from sklearn.model_selection import cross_val_score
svc = SVC(kernel='linear')
score = cross_val_score(svc , X ,y , cv=10 , scoring ="accuracy")
print(score)
```

```python
print(score.mean())
```

```python
from sklearn.model_selection import cross_val_score
svc = SVC(kernel="rbf")
score = cross_val_score(svc, X , y ,cv = 10 , scoring = "accuracy")
print(score)
```

```python
from sklearn.model_selection import cross_val_score
svc = SVC(kernel = "poly")
score = cross_val_score(svc , X, y, cv= 10 , scoring = "accuracy")
print(score)
```

```python
print(score.mean())
```

```python
C_range = list(range(1,26))
acc_score = []
for c in C_range:
    svc = SVC(kernel="linear" , C=c)
    score = cross_val_score(svc , X, y , cv = 10 , scoring= "accuracy")
    acc_score.append(score.mean())
print(acc_score)
```

```python
# libraries and values are already definded
plt.plot(C_range, acc_score)
plt.xticks(np.arange(0,27,2))
plt.xlabel("Value of C for SVC")
plt.ylabel("Cross-Validation Accuracy")
plt.show()
```

```python
C_range = list(np.arange(0.1 , 6, 0.1))
acc_score = []
for c in C_range:
    svc = SVC(kernel='linear', C=c)
    score = cross_val_score(svc , X, y , cv= 10 , scoring="accuracy")
    acc_score.append(score.mean())
print(acc_score)  
```

```python
C_range = list(np.arange(0.1, 6 ,0.1))
plt.plot(C_range, acc_score)
plt.xticks(np.arange(0.0, 6,0.5))
plt.xlabel("Value of C for SVC")
plt.ylabel("Cross-Validation Accuracy")
plt.show()
```

```python
gamma_range=[0.0001,0.001,0.01,0.1,1,10,100]
acc_score = []
for g in gamma_range:
    svc = SVC(kernel ='rbf' , gamma = g)
    score = cross_val_score(svc, X, y , cv = 10 , scoring = 'accuracy')
    acc_score.append(score.mean())
print(acc_score)
```

```python
gamma_range=[0.0001,0.001,0.01,0.1,1,10,100]
plt.plot(gamma_range, acc_score)
plt.xticks(np.arange(0.0001, 100, 5))
plt.xlabel("Value of gamma for SVC")
plt.ylabel("Cross-validation Accuracy")
plt.show()
```
